```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

# Ödev 3 - Soru 7 

```python
pi = np.pi
w = np.linspace(0, 2*pi,50)
H = np.zeros(len(w),dtype = complex)
H_amp = np.zeros(len(w))
H_phase = np.zeros(len(w))

for i in range(len(w)):
    H[i] = (1 - (0.45*(np.exp(-(0 + 1j)*w[i]))/(1-0.8*(np.exp(-(0 + 1j)*w[i])))))
    H_amp = np.abs(H)
    H_phase = np.angle(H)



plt.plot(w/pi, H_amp)
plt.title('$|X(e^{j\omega})|$')
plt.xlabel('w(x$\pi$)')

plt.figure()

plt.plot(w/pi, H_phase)
plt.title('$ \measuredangle X(e^{j\omega})$')
plt.xlabel('w(x$\pi$)')


```

# Ödev 3 - Soru 8

```python
pi = np.pi
w = np.linspace(0, 2*pi,100)
X = np.zeros(len(w),dtype = complex)
X_amp = np.zeros(len(w))
X_phase = np.zeros(len(w))

for i in range(len(w)):
    X[i] = (1/(1-(-0.5)*np.exp(-1j*w[i])))
    
X_amp = np.abs(X)
X_phase = np.angle(X)



plt.plot(w/pi, X_amp)
plt.title('$|X(e^{j\omega})|$')
plt.xlabel('w(x$\pi$)')

plt.figure()

plt.plot(w/pi, X_phase)
plt.title('$ \measuredangle X(e^{j\omega})$')
plt.xlabel('w(x$\pi$)')
```

# Ödev 3 - Soru 9

```python
pi = np.pi
w = np.linspace(0, 2*pi,100)
H = np.zeros(len(w),dtype = complex)
H_amp = np.zeros(len(w))
H_phase = np.zeros(len(w))

for i in range(len(w)):
    H[i] = ((1 + np.exp(-2j*w[i]))/(1 - 0.8*np.exp(-1j*w[i])))
    
H_amp = np.abs(H)
H_phase = np.angle(H)



plt.plot(w/pi, H_amp)
plt.title('$|H(e^{j\omega})|$')
plt.xlabel('w(x$\pi$)')

plt.figure()

plt.plot(w/pi, H_phase)
plt.title('$ \measuredangle H(e^{j\omega})$')
plt.xlabel('w(x$\pi$)')
```
