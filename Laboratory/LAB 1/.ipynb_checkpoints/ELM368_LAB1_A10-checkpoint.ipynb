```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
from scipy.fftpack import fft , ifft
```

## SORU 1-a)

```python
# EKOK(1/10,1/15) = 1/5 = T 
pi = np.pi
t = np.linspace(0,2/5,1000)

x = 2*np.sin(20*pi*t - pi/5) + np.cos(30*pi*t - 3*pi/5)
plt.plot(t,x)
plt.xlabel('t',  fontsize='20')  
plt.ylabel('x(t)',  fontsize='20') 
plt.title(r'$x(t) = 2sin(20\pi t - \pi / 5) + cos(30\pi t - 3\pi /5)  T = 0.2s $', fontsize='16') 

```

## Soru 1-b)

```python
# EKOK(1/5,2/15) = 2/5 = T 
pi = np.pi
t = np.linspace(0,4/5,1000)

x = 2*np.sin(10*pi*t - pi/5) + np.cos(15*pi*t - 3*pi/5)
plt.plot(t,x)
plt.xlabel('t',  fontsize='20')  
plt.ylabel('x(t)',  fontsize='20') 
plt.title(r'$x(t) = 2sin(10\pi t - \pi / 5) + cos(15\pi t - 3\pi /5)  T = 0.4s $', fontsize='14') 
```

```python
# EKOK(7/2,7/3) = 7 = N
pi = np.pi
n = np.linspace(0,13,14)
x = np.zeros(len(n))

x = 2*np.sin(4*pi*n/7 - pi/5) + np.cos(6*pi*n/7 - 3*pi/5)
plt.stem(n,x,use_line_collection=True)
plt.xlabel('n',  fontsize='20')  
plt.ylabel('x[n]',  fontsize='20') 
plt.title(r'$x[n] = 2sin(\frac{4\pi n}{7} - \pi / 5) + cos(\frac{6\pi n}{7} - 3\pi /5)  N = 7 sample $', fontsize='14') 
```

## SORU 2-h)

```python
pi = np.pi
w = np.linspace(0, 2*pi, 100)
M = 3
Habs = np.abs((1/(2*M+1)) * np.sin(w*(2*M + 1)/2)/(np.sin(w/2)))
Hangle = np.angle((1/(2*M+1)) * np.sin(w*(2*M + 1)/2)/(np.sin(w/2)))

plt.plot(w/pi,Habs)
plt.xlabel('w(xpi)',  fontsize='20')  
plt.ylabel('H3(w)',  fontsize='20') 
plt.title(r'$H_3(e^{jw}) Genlik $', fontsize='14') 

plt.figure()
plt.plot(w/pi,Hangle/pi)
plt.xlabel('w(xpi)',  fontsize='20')  
plt.ylabel('H3(w)',  fontsize='20') 
plt.title(r'$H_3(e^{jw}) Faz $', fontsize='14') 

M = 9
Habs = np.abs((1/(2*M+1)) * np.sin(w*(2*M + 1)/2)/(np.sin(w/2)))
Hangle = np.angle((1/(2*M+1)) * np.sin(w*(2*M + 1)/2)/(np.sin(w/2)))

plt.figure()
plt.plot(w/pi,Habs)
plt.xlabel('w(xpi)',  fontsize='20')  
plt.ylabel('H3(w)',  fontsize='20') 
plt.title(r'$H_3(e^{jw}) Genlik$', fontsize='14') 

plt.figure()
plt.plot(w/pi,Hangle/pi)
plt.xlabel('w(xpi)',  fontsize='20')  
plt.ylabel('H9(w)',  fontsize='20') 
plt.title(r'$H_9(e^{jw}) Faz $', fontsize='14') 
```

## SORU 2-I)

```python

pi = np.pi
M = 3
n = np.linspace(0,40,41) 
x = np.zeros(len(n))
h = np.zeros(len(n))

for k in range(len(n)):
    if(n[k]<= M and n[k]>= -M ):
        h[k] = (1/(2*M+1))

x = np.cos(pi*n/9) + np.cos(7*pi*n/9)
L = len(x) + len(h) - 1

y = signal.convolve(x,h,'same')
plt.plot(n,y)
plt.xlabel('n',  fontsize='20')  
plt.ylabel('y[n]',  fontsize='20') 
plt.title(r'$Çıkış Sinyali $', fontsize='14') 

M = 9

for k in range(len(n)):
    if(n[k]<= M and n[k]>= -M ):
        h[k] = (1/(2*M+1))

x = np.cos(pi*n/9) + np.cos(7*pi*n/9)
L = len(x) + len(h) - 1
plt.figure()
y = signal.convolve(x,h,'same')
plt.plot(n,y)
plt.xlabel('n',  fontsize='20')  
plt.ylabel('y[n]',  fontsize='20') 
plt.title(r'$Çıkış Sinyali $', fontsize='14') 
```
