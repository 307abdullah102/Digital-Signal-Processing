```python
import numpy as np 
import matplotlib.pyplot as plt 
import sys 
import math
```

```python
# x1[n] - x1(t) = 7cos(200*pi*t) örneklenmesi

Fs=400 # örnekleme frekansını
Ts=1/Fs # örnekleme periyodu

# pi*N1/20 = 2pi*k ==> N1 = 20k, k = 1 için N1 = 20
# 3pi*N2/40 = 2pi*k ==> 3N2 = 80k, k = 3 için N1 = 80

N=80 # örnek sayısı
n=np.arange(0.,2*N) # 2 periyot çizdirme

pi = np.pi
xn = 4 + 3*np.sin(n*pi/20) + 2*np.cos(3*n*pi/40 - 3*pi/4) 

plt.figure(figsize=(12,4))
plt.stem(n,xn)
plt.title("Örnekleme sonucu oluşan işaret", ) 
plt.xlabel("N(sample)")
plt.ylabel("x[n]")
plt.show()
```

## B)

```python
w = np.arange(-np.pi, np.pi, 2*np.pi/N) # omega ekseninin -pi ile +pi arasında tanımlanması
xw = np.fft.fftshift(np.fft.fft(xn,N)/N) # ayrık zamanlı işaretin Fourier transformu
```

```python
plt.figure(figsize=(12,4))
plt.stem(w/np.pi,abs(xw)) # X(w) işaretinin grafiğinin çizdirilmesi
plt.xlabel("$\omega$ / $\pi (rad)$") # grafiğin x ekseninin isimlendirilmesi
plt.title("$X(\omega)$ (Genlik)") # grafiğin y ekseninin isimlendirilmesi
plt.show()

plt.figure(figsize=(12,4))
plt.stem(w/np.pi,np.angle(xw)) # X(w) işaretinin grafiğinin çizdirilmesi
plt.xlabel("$\omega$ / $\pi (rad)$") # grafiğin x ekseninin isimlendirilmesi
plt.title("$X(\omega)$ (Faz)") # grafiğin y ekseninin isimlendirilmesi
plt.show()
```

## C)

### x[n] = K + A1sin(w1n + a1) + A2cos(w2n + a2) 
### B şıkkındaki genlik grafiğinde, w = 0'da görülen alanı 4 olan dirac ters fourierde xn sinyalinin katsayısını ifade eder. Bu durumda K = 4 bulunur. simetrik 2 adet 1 alanlı diraclar o noktadaki faz değerine bakıldığında cos sinyalini ifade eder. Bu durumda A2 = 2*1, a2 = 2.35 = 3pi/4 olduğu görülür. A1 = 1.5*2 = 3 olduğu görülür. a1 = 1.5 = pi/2 bu faz farkı sin ve cos sinyalleri arasındaki pi/2'lik faz farkından gelmektedir. 0-0.25 arası 10 bölüm olduğundan 0.050pi = 1/20pi = w1 bulunur.0.075pi = 3/40pi = w2
