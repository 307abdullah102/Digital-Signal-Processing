```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = (3 - 4j)
b = (1 + 2j)
C = a + b
D = a / b
Amp_C= math.sqrt(pow(C.real,2) + pow(C.imag,2))
Phase_C = math.atan(C.imag/C.real)

Amp_D= math.sqrt(pow(D.real,2) + pow(D.imag,2))
Phase_D = math.atan(D.imag/D.real)

# 1- Gerçek bileşenleri bastırma
print("C kompleks sayısının gerçek bileşeni: {0}".format(C.real))
print("D kompleks sayısının gerçek bileşeni: {0}".format(D.real))
print("\n")

# 2- Sanal bileşenleri bastırma
print("C kompleks sayısının sanal bileşeni: {0}".format(C.imag))
print("D kompleks sayısının sanal bileşeni: {0}".format(D.imag))
print("\n")

# 3- Genlik bileşenleri bastırma
print("C kompleks sayısının genlik bileşeni: {0}".format(Amp_C))
print("D kompleks sayısının genlik bileşeni: {0}".format(Amp_D))
print("\n")

# 4- Faz bileşenleri bastırma
print("C kompleks sayısının faz bileşeni: {0}".format(Phase_C))
print("D kompleks sayısının faz bileşeni: {0}".format(Phase_D))
print("\n")

```

# SORU - 2

```python
pi = np.pi
# w = 2*pi*f, f = 1/7 , T = 1/f , T = 7, 2T = 14 
w = (2 * pi / 7)
a = (- pi / 5)
# 2T bastırabilmek için 14.01'e kadar
#(arange fonksiyonunda üst sınır dahil değil)
n = np.arange(0,  14 + 0.1, 1) 
x = np.zeros(len(n))

for i in range(len(n)):
    x[i] = 2 * np.sin((w * i) + a)
    
plt.stem(n, x,  use_line_collection=True)

# x ekseninde değerleri her adımda bastırmak için 
plt.xticks(np.arange(n[0],n[-1],1))     

plt.xlabel('n',  fontsize='20')  
plt.ylabel('x[n]',  fontsize='20')  
plt.title(r'$x[n] = 2sin(\frac{2\pi}{7} n - \frac{\pi}{5})$', fontsize='20') 


plt.show()
```

## 3.1)

```python
pi = np.pi
n = np.arange(0,  15 + 0.1, 1)
x = np.zeros(len(n))

for i in range(len(n)):
    x[i] = np.exp(pi/5)
    
plt.stem(n, x,  use_line_collection=True)


plt.xticks(np.arange(n[0],n[-1],1))     

plt.xlabel('n',  fontsize='20')  
plt.ylabel('x[n]',  fontsize='20')  
plt.title('$x_1[n] = e^{(\pi/5)}$', fontsize='20') 


plt.show()
```

##  $x_1[n]$ sinyali n'den bağımsız bir sinyal olduğundan n'in değişen değerleri için sabit bir değer almaktadır.

## 3.2)

```python
pi = np.pi
n = np.arange(0,  15 + 0.1, 1)
x = np.zeros(len(n))
xImag = np.zeros(len(n))
for i in range(len(n)):
    x[i] = (np.exp((0 + 1j)*pi/5)).real
    xImag[i] = (np.exp((0 + 1j)*pi/5)).imag

    


# print Real part of signal    
plt.subplot(2,1,1)
plt.stem(n, x,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Re $x_2[n]$',  fontsize='20')  

# Title of 2 plots 
plt.title(r'$ x_2[n] = e^{(j*\pi/5)}$', fontsize='20') 

#print imag part of signal
plt.subplot(2,1,2)
plt.stem(n, xImag,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Im $x_2[n]$',  fontsize='20')  



plt.show()
```

##  $e^{j*\frac{\pi}{5}} = cos(\frac{\pi}{5}) + j*sin(\frac{\pi}{5})$ sinyaline karşılık gelir sinyal n'den bağımsız olduğu için n'in değişen değerlerine karşılık sabit $x_2[n]$ değerleri elde edilir.  Karmaşık bir sinyal olmasından dolayı ise sinyalin gerçek ve sanal olmak üzere iki ayrı grafiği elde edilir.

## 3.3)

```python
pi = np.pi
n = np.arange(0,  15 + 0.1, 1)
x = np.zeros(len(n))
xImag = np.zeros(len(n))

for i in range(len(n)):
    x[i] = (np.exp((0 + 1j)*i*pi/5)).real
    xImag[i] = (np.exp((0 + 1j)*i*pi/5)).imag

      
    
# print Real part of signal  
plt.subplot(2,1,1)
plt.stem(n, x,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Re x[n]',  fontsize='20')  

# Title of 2 plots 
plt.title('$x_3[n] = e^{(j*n*\pi/5)}$', fontsize='20') 

#print imag part of signal
plt.subplot(2,1,2)
plt.stem(n, xImag,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Im x[n]',  fontsize='20') 

plt.show()
```

## Hem n'e bağımlı hem de kompleks olan $x_3[n]$ sinyali $ \omega = 2 \pi f = \pi / 5 \Longrightarrow f = \frac{1}{10} \Longrightarrow T = 10 $ periyodik bir bir sinyaldir.

## 3.4)

```python
pi = np.pi
n = np.arange(0,  15 + 0.1, 1)
x = np.zeros(len(n))
xImag = np.zeros(len(n))

for i in range(len(n)):
    x[i] = (np.cos(pi/5) + (0 + 1j)*np.sin(pi/5)).real
    xImag[i] = (np.cos(pi/5) + (0 + 1j)*np.sin(pi/5)).imag

    
# print Real part of signal   
plt.subplot(2,1,1)
plt.stem(n, x,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Re x[n]',  fontsize='20') 

# Title of 2 plots 
plt.title(r'$x_4[n] = cos(\frac{\pi}{5}) + j*sin(\frac{\pi}{5})$', fontsize='20') 

#print imag part of signal
plt.subplot(2,1,2)
plt.stem(n, xImag,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Im x[n]',  fontsize='20')  
 


plt.show()
```

## Şekilde görülen grafiğin 3.2 grafiği ile aynı olduğu gözlemlenmiştir. Bu şart euler formülü ile açıklanır: $ e^{j* \alpha } = cos( \alpha ) + j*sin( \alpha ) $
## Euler formülü gereği $ x_2[n] = x_4[n] $ eşitliği sağlanır.

## 3.5)

```python
pi = np.pi
n = np.arange(0,  15 + 0.1, 1)
x = np.zeros(len(n))
xImag = np.zeros(len(n))

for i in range(len(n)):
    x[i] = (np.cos(i*pi/5) + (0 + 1j)*np.sin(i*pi/5)).real
    xImag[i] = (np.cos(i*pi/5) + (0 + 1j)*np.sin(i*pi/5)).imag

    
# print Real part of signal      
plt.subplot(2,1,1)  
plt.stem(n, x,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Re x[n]',  fontsize='20')  

# Title of 2 plots 
plt.title(r'$x_5[n] = cos(\frac{\pi}{5}n) + j*sin(\frac{\pi}{5}n)$', fontsize='20') 

#print imag part of signal
plt.subplot(2,1,2)  
plt.stem(n, xImag,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Im x[n]',  fontsize='20')  

plt.show()
```

## Aynı bağlamda 3.3 ve 3.5 sinyallerinde eşlik söz konusudur. Bu eşlik euler formülü ile açıklanır: $ e^{j* \alpha } = cos( \alpha ) + j*sin( \alpha ) $
## Euler formülü gereği $ x_3[n] = x_5[n] $ eşitliği sağlanır. Sinyalin eşliğinden dolayı 3.3'teki sinyal periyodu (T = 10) değeri 3.5 sinyali için de geçerlidir.

```python
pi = np.pi
n = np.arange(-10,  4 + 0.1, 1)
x = np.zeros(len(n))
xImag = np.zeros(len(n))

for i in range(len(n)):
    x[i] = (np.cos((i-10)*pi/5) + (0 + 1j)*np.sin((i-10)*pi/5)).real
    xImag[i] = (np.cos((i-10)*pi/5) + (0 + 1j)*np.sin((i-10)*pi/5)).imag

    
    
# print Real part of signal
plt.subplot(2,1,1)    
plt.stem(n, x,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Re x[n]',  fontsize='20')

# Title of 2 plots 
plt.title(r'$x_6[n] = cos(\frac{\pi}{5}n) + j*sin(\frac{\pi}{5}n)$', fontsize='20') 

#print imag part of signal
plt.subplot(2,1,2)    
plt.stem(n, xImag,  use_line_collection=True)
plt.xticks(np.arange(n[0],n[-1],1))     
plt.xlabel('n',  fontsize='20')  
plt.ylabel('Im x[n]',  fontsize='20')  


plt.show()
```
