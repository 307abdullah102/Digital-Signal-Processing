## a) 2.7 - (a) sinyalinin çizimi 

Kütüphanelerin Çağrılması:

```python
import numpy as np
import matplotlib.pyplot as plt 
import math
#%matplotlib notebook
```

```python
pi = np.pi

n = np.arange(-24,24,1)
y = np.zeros(len(n))
y_imag = np.zeros(len(n))
for i in range(len(n)):
    y[i] = (np.exp(((0+1j)*pi*(i-24)/6))).real
    y_imag[i] = (np.exp(((0+1j)*pi*(i-24)/6))).imag
    
# Real kısmı çizdirme
plt.subplot(2,1,1)
plt.stem(n, y,  use_line_collection=True)
# x eksenindeki sayıların karmaşıklığını engellemek için her 3 birimde bir yazdırmayı sağlar.
plt.xticks(np.arange(n[0],n[-1],3))     
# Periyot aralığı el ile girilerek grafik üzerinde bir periyodun mavi ile işaretlenmesini sağlar.
plt.axvspan(0, 12, color="blue", alpha=0.3)
plt.xlabel('n',  fontsize='20')  
plt.ylabel('y[n]',  fontsize='20')  

# title of plot
plt.title('y[n] = $e^{(j*\pi*n/6)}$', fontsize='20') 

# Imaginary kısmı çizdirme
plt.subplot(2,1,2)
plt.stem(n, y_imag,  use_line_collection=True)
# x eksenindeki sayıların karmaşıklığını engellemek için her 3 birimde bir yazdırmayı sağlar.
plt.xticks(np.arange(n[0],n[-1],3))     
# Periyot aralığı el ile girilerek grafik üzerinde bir periyodun mavi ile işaretlenmesini sağlar.
plt.axvspan(0, 12, color="blue", alpha=0.3)
plt.xlabel('n',  fontsize='20')  
plt.ylabel('y[n]',  fontsize='20')  


plt.show()
```

### a) şıkkında elde edilen grafikte y sinyali 4 periyot halinde çizdirilmiştir. Mavi bölge grafiğin bir periyodudur.(N = 12) bu değer, el ile hesaplanan değer ile eşdeğerdir.

## 9-a)

```python


n = np.arange(0,30,1)
y = np.zeros(len(n))
A = 5
a = 0.8
for i in range(len(n)):
    y[i] = A * (a ** i)
    

plt.stem(n, y, use_line_collection=True)
#plt.legend(loc='upper right')
plt.annotate('A=5, a=-0.8', xy=(1, 2), xytext=(20,4.5),fontsize='15')
plt.xticks(np.arange(n[0],n[-1],2)) 


plt.xlabel('n',  fontsize='20')  
plt.ylabel('x[n]',  fontsize='20')  
plt.title('x[n] = $A*a^{n}$', fontsize='20') 


plt.show()
```

### Yukarıdaki grafikte 0 < a < 1 değerleri için grafikteki değerlerin n değerinin artışıyla azaldığı gözlenir. Bu durum a^n teriminden dolayı ve a değer aralığından dolayı beklenen durumdur. n = 0 için x[n] = A değeri alır.

# ÖDEV - 1 -- SORU - 10

## 10-a)

```python
pi = np.pi
n = np.arange(0,6*pi + 0.001,pi/8)
y = np.zeros(len(n))
w = 0.1*pi
a = pi/4
for i in range(len(n)):
    y[i] = np.sin((w * i) + a)
    

plt.stem(n, y, use_line_collection=True)
#plt.legend(loc='upper right')
plt.annotate('$ w = 0.1*\pi, a=\pi / 4$', xy=(0, 1), xytext=(20,0.85),fontsize='15')
plt.xticks(np.arange(n[0],n[-1],2)) 


plt.xlabel('n',  fontsize='20')  
plt.ylabel('x[n]',  fontsize='20')  
plt.title('x[n] = $sin(\omega*n + a)$', fontsize='20') 


plt.show()
```

### sin fonksiyonu n = 0'da herhangi bir faz kayması olmadıkça 0 değeri almalıdır. Burada a değeri faz kaymasını belirler. sinyalin belirli noktada hangi değeri alacağını belirlemektedir. Periyodu değiştirecek bir etkide bulunmaz. Böylece devamında w değeri değişse dahi x[0] = 1 değeri a değeri değişmedikçe sağlanır.
